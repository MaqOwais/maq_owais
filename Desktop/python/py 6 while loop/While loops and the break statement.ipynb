```python
# for loops are not the only way to use loops in python the other one is while loops 
total = 0
for i in range (1,5):
    total += i
print (total)
```

```python
# the other one is while loop same as above in different way
total1=0
j=1
# (a) says that check weather the j is < 5 as j is equal to 1 as it is true then do the following i.e add j to total1 
# and so on and when j is taken as 5 then while statenment is not satisfied hence loop will break
while j < 5:#..(a)
    total1 += j # shortcut as total1 = total1 + j
    j += 1
print (total1)
```

```python
# the advantage of while loop over forloop is (depends on the situation)
w = [5,3,1 , 4, 4, 4, 3, 3]

total2 = 0 
i = 0 
while i < len(w) and w[i] > 0:
    total2 += w[i]
    i += 1
print(total2)
```

```python
# the above one can be done by for loop
w = [5,3,1,-1]
tot = 0 
for ele in w :
    if ele <= 0 :
        break # is a predefined fn used to break the statement
    tot += ele
print(ele)
```
