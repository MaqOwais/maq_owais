```python
a= ['bananna', 'apple' , 'micro']
```

```python
# to print each element 
print(a[0])
print(a[1])
print(a[2])
```

```python
# it could to complex for more no. of terms in the array So....we use for loop
for element in a :
    print(element)
# the block (code) says 'for each element in 'a' do the following' (i.e. printing each element in it)
```

```python
# we can also have multiple statements in the above block as well
for element in a :
    print(element)
    print(element)
```

```python
# here we can replace the 'element' as any character as a variable
for e in a :
    print(e)
    print(e)
```

```python
# here we can sum up the no. in the array
# here 'total' is variable
b = [ 20 ,1 , 23]
total = 0
for e in b:
    total = total + e
print(total)
```

```python
# what if we want to find the sum of 1-4
# ans : we are making a new list [1,2,3,4] of no.s but it could be done in the better way
# i.e. to use the range fn in python so we write
range(1,5)
# this means it is create a range of no.s starting at one to 5 but not including the no. 5
#       i.e [1,5)
# n this method is better for finding the sum from 1 to 5 since we no need to write [1,2,...]
# is sort of like a list can also convert to list as list(range(1,5))
c=list(range(1,5))
print (c)
```

```python
range(1,5)
c=list(range(1,5))
print (c)
```

```python
d = range(1,5)
print (d)
```

```python
# we can use the 'range ' for 'for loop' So..
for i in range (1,5):
    print(i)
```

```python
# we can have the sum of it as the above  code
toal = 0
for i in range (1,5):
    toal = toal + i
# we can also have short cut as toal += i
print (toal)
```
